# assigning the features and target variable into x and y in train data

```python
x=df1 
y=df1.pop("Average Flight Fare")
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25,random_state=9)
```

# Rescaling the Features

```python
# StandardScaler follows Standard Normal Distribution (SND).
# Therefore, it makes mean = 0 and scales the data to unit variance. 

from sklearn.preprocessing import StandardScaler
st_x=StandardScaler()
x_train=st_x.fit_transform(x_train)
x_test=st_x.transform(x_test)
```

# Fitting the model

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(x_train,y_train)
```

```python
print('Coefficients: ', lm.coef_)
```

```python
#make prediction
pred=lm.predict(x_test)
```

```python
from sklearn import metrics
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Calculate evaluation metrics
mae = mean_absolute_error(y_test,pred)
mse = mean_squared_error(y_test,pred)
r_squared = r2_score(y_test,pred)
rmse = np.sqrt(mse)
 
# Print the evaluation metrics
print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
print("R-squared (RÂ²):", r_squared)
print("Root Mean Squared Error (RMSE):", rmse)
```

# Residual Analysis

```python
y_train_price = lm.predict(x_train)
```

```python
# Plot the histogram of the error terms
fig = plt.figure()
sns.distplot((y_train - y_train_price), bins = 20)
fig.suptitle('Error Terms', fontsize = 20)                  # Plot heading 
plt.xlabel('Errors', fontsize = 18)                         # X-label
```

<!--  Error Terms are normally distributed - hence prediction can be made -->
